# Question 5: How safe is it to cycle in London?

## Investigation into road accidents/casualties

Data from GOV.UK website: Reported road accidents, vehicles and casualties tables for Great Britain. This can be found on: https://www.gov.uk/government/statistical-data-sets/reported-road-accidents-vehicles-and-casualties-tables-for-great-britain#casualties-in-reported-road-accidents-ras30. The file used was: "RAS30043: Casualties by severity, region, local authority and road user type: England." This data in this file was extracted, rearranged, and converted into a csv file in Excel and then analysed below.

```python
# Import modified csv file
cyclist_casualties_df = pd.read_csv("Cyclist casualties.csv")
```

### Explore and visualise data

```python
# Cycling casualties by year - all casualty types
sns.set(rc = {'figure.figsize':(25,10)}, font_scale = 10)
sns.set_theme(style="ticks")
y = sns.lineplot(data = cyclist_casualties_df, x = 'Year', y = 'Number', hue="Casualty Type", ci = None)
y.set_title('Cycling casualties in London by year', fontsize = 30)
y.set_xlabel('Year', fontsize = 20)
y.set_ylabel('Count', fontsize = 20)
plt.tick_params(axis='both', labelsize=20)
```

```python
# Subset by 'All casualties'
cyclist_casualties_all_df = cyclist_casualties_df[cyclist_casualties_df['Casualty Type'] == 'All casualties'].reset_index(drop = True)
cyclist_casualties_all_df
```

```python
# Cycling casualties by year - all casualty types
sns.set(rc = {'figure.figsize':(25,10)}, font_scale = 10)
sns.set_theme(style="ticks")
y = sns.lineplot(data = cyclist_casualties_all_df, x = 'Year', y = 'Number', ci = None)
y.set_title('Cycling casualties in London by year - All casualties', fontsize = 30)
y.set_xlabel('Year', fontsize = 20)
y.set_ylabel('Count', fontsize = 20)
plt.tick_params(axis='both', labelsize=20)
```

```python
# Subset by 'Killed or seriously injured'
cyclist_casualties_KIS_df = cyclist_casualties_df[cyclist_casualties_df['Casualty Type'] == 'Killed or seriously injured'].reset_index(drop = True)
cyclist_casualties_KIS_df
```

```python
# Cycling casualties by year - killed or seriously injured
sns.set(rc = {'figure.figsize':(25,10)}, font_scale = 10)
sns.set_theme(style="ticks")
y = sns.lineplot(data = cyclist_casualties_KIS_df, x = 'Year', y = 'Number', ci = None)
y.set_title('Cycling casualties in London by year - Killed or seriously injured', fontsize = 30)
y.set_xlabel('Year', fontsize = 20)
y.set_ylabel('Count', fontsize = 20)
plt.tick_params(axis='both', labelsize=20)
```
