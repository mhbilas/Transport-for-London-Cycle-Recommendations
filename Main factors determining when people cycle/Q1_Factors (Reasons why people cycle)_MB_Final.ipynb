## Explore and visualise data

```python
# Reasons for cycling by year
sns.set(rc = {'figure.figsize':(25,10)}, font_scale = 10)
sns.set_theme(style="ticks")
y = sns.lineplot(data = reasons_cycle_df, x = 'Year', y = 'Percentage', hue="Reason", ci = None)
y.set_title('Reasons for cycling in London', fontsize = 30)
y.set_xlabel('Year', fontsize = 20)
y.set_ylabel('Percentage', fontsize = 20)
plt.tick_params(axis='both', labelsize=20)
```

Observations: There are more people who cycle for travel than for leisure. There is an overall decrease in cycling from 2016-2017. There is a gradual increase in the number of people who cycle for travel from 2016, onward, with a dip from 2019-2020. Meanwhile, we see a spike in those who cycle for leisure from 2019-2020. This trend makes sense in light of the pandemic.

```python
# Cycle rates
sns.set(rc = {'figure.figsize':(25,10)}, font_scale = 10)
sns.set_theme(style="ticks")
y = sns.barplot(data = reasons_cycle_df, x = 'Rate', y = 'Percentage', hue="Reason", ci = None)
y.set_title('Cycling rates', fontsize = 30)
y.set_xlabel('Rate', fontsize = 20)
y.set_ylabel('Percentage', fontsize = 20)
plt.tick_params(axis='both', labelsize=20)
```

Observations: People who cycle for travel cycle more frequently than those who cycle for leisure, although both follow a similar trend.
